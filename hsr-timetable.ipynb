```python
import requests
from bs4 import BeautifulSoup
import  pandas as pd
```

```python
form_data = {
"SearchType": 'S',
"StartStation":"977abb69-413a-4ccf-a109-0272c24fd490", 
"EndStation":"e6e26e66-7dc1-458f-b2f3-71ce65fdc95f",  
"DepartueSearchDate":"2019/08/23",
"DepartueSearchTime":"11:00" 
}
response_post = requests.post("https://www.thsrc.com.tw/tw/TimeTable/Search", data = form_data)
soup_post = BeautifulSoup(response_post.text,"lxml")
data = response_post.json()
```

```python
df = pd.DataFrame(columns = ["車次", "出發時間", "抵達時間", "行車時間"])
times = []
columns = ['TrainNumber', 'DepartureTime', 'DestinationTime', 'Duration']
for d in data['data']['DepartureTable']['TrainItem']:
    times.append([d[c] for c in columns])
i = 1
for t in times:
    df.loc[i] = (t)
    i += 1
df
```
